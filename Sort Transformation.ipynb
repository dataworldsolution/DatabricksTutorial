## Sorting Data in PySpark using Databricks

- Sorting data in ascending and descending order

- Finding top N records
 
- Sorting using multiple columns
 
- Sorting with different order for each column

**Step 1: Read and Explore the Data**

```python
df = spark.read.table('ytdemo.data.global_superstore')
display(df)
```

## Basic Sorting (Ascending)

```python
df.sort('Sales').display()
```

### Sort Sales in Descending Order

```python
df.sort('Sales', ascending = False).display()
```

### Find Top 10 Sales Records

```python
df.sort('Sales',ascending = False).limit(10).display()
```

### Sort Using Multiple Columns

```python
df.sort('State','Sales' , ascending = False).select ('State','Sales').display()
```

### Sort with Different Order for Each Column

```python
df.sort(df.Order_Date.asc(),df.Sales.desc()).select('Order_Date','Sales').display()
```

```python
df.sort(df['Order_Date'].asc(),df['Sales'].desc()).select('Order_Date','Sales').display()
```
