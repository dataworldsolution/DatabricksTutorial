```python
df = spark.read.csv('/Volumes/ytdemo/data/demodata/GlobalSuperstore.csv', 
                    header=True, 
                    inferSchema=True)
display(df)
```

```python
from pyspark.sql.functions import col , round , greatest , least
```

```python
df.columns
```

```python
df = df.withColumn('Sales_Adjsuted',col('Sales')*1.1)
df.select('Sales','Sales_Adjsuted').show()
```

```python
from pyspark.sql.functions import sum , avg
```

```python
total_sales = df.agg(sum('Sales').alias('Total_Sales'))
total_sales.show()
```

```python
total_qty = df.agg(sum('Quantity').alias('Total_Quantity'))
total_qty.show()
```

```python
avg_profit = df.agg(avg('Profit').alias('Average_Profit'))
avg_profit.show()
```

```python
df = df.withColumn('Profit_Margin',round(col('Profit')/col('Sales'),2))
df.select('Sales','Profit','Profit_Margin').show()
```

## Using greatest() and least() Functions

```python
df = df.withColumn(
    "Max_Value", 
    greatest(col("Sales"), col("Profit"), col("Discount"))
)

df.select("Sales", "Profit", "Discount", "Max_Value").show(5)
```

```python
df = df.withColumn(
    "Min_Value", 
    least(col("Sales"), col("Profit"), col("Discount"))
)

df.select("Sales", "Profit", "Discount", "Min_Value").show(5)
```
