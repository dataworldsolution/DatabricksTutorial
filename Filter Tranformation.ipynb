## Filter Transformation

DataFrame.filter(condition)

Filters rows using the given condition.

where() is an alias for filter().



**Loading and Visualizing Data**

```python
df = spark.read.table('ytdemo.data.global_superstore')
display(df)
```

**Basic Filtering**

```python
# pythonic way

df.filter(df.Category == 'Furniture').display()

# sql way

df.filter("Category == 'Furniture'").display()
```

**Exploring Unique Values**

```python
df.select('Category').distinct().show()
```

**Multiple Conditions (OR Logic)**

```python

#python way
df.filter((df.Category == 'Furniture') | (df.Category == 'Technology')).display()

#sql
df.filter("Category == 'Furniture' OR Category == 'Technology'").display()
```

**Refining Results (AND Logic)**

```python
# pythonic way
df.filter((df.Category == 'Furniture') & (df.Sub_Category == 'Tables')).display()

# SQL Style
df.where("Category = 'Furniture' and Sub_Category = 'Tables'").display()
```

**Filtering by dates**

```python
from pyspark.sql.functions import col
df.filter(df.Order_Date >= '2015-01-01').display()
df.filter(col('Order_date').between('2015-01-01','2015-12-31')).display()
df.filter("order_date between '2015-01-01' and  '2015-12-31'").display()
```

**Filtering with math**

```python
# pythonic
df.filter((df.Profit/df.Sales)>0.10).display()

# SQL Style:
df.filter("(Profit/Sales)>0.10").display()
```
