### Aggregation Function

**_Aggregation functions that are commonly used
Such As_**
- sum
- avg
- count
- min
- max
- mean

```python
df = spark.read.table('ytdemo.data.global_superstore')
df.display()
```

```python
type(df.groupBy("region"))
```

```python
from pyspark.sql.functions import sum , avg, max, min
```

```python
df.groupBy("region").sum("Sales").avg("Sales").display()
```

```python
df.groupBy("region").avg("Sales").display()
```

### Using agg() for Multiple Aggregations

```python
df.groupBy("region").agg(
                          sum("Sales").alias("Tatal Sales"),
                          avg("Sales").alias("Average Sales"),
                          max("Sales").alias("Maximum Sales"),
                          min("Sales").alias("Minimum Sales")
                        ).display()
```

### Grouping by Multiple Columns

```python
df.groupBy("region","Segment").agg(
                          sum("Sales").alias("Tatal Sales"),
                          avg("Sales").alias("Average Sales"),
                          max("Sales").alias("Maximum Sales"),
                          min("Sales").alias("Minimum Sales")
                        ).display()
```
