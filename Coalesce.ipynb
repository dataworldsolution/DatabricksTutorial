## PySpark coalesce() Function in Databricks (Real-World Example)

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

schema = StructType([
    StructField("order_id", IntegerType(), False),
    StructField("customer_name", StringType(), True),
    StructField("email", StringType(), True),
    StructField("phone_number", StringType(), True),
    StructField("coupon_code", StringType(), True),
    StructField("shipping_address", StringType(), True),
    StructField("order_amount", DoubleType(), True)
])

data = [
    (1001, "Amit Sharma", "amit@gmail.com", "9876543210", "NEW10", "Delhi", 1500.50),
    (1002, "Priya Verma", None, None, None, "Mumbai", 2450.00),
    (1003, "Rahul Singh", "rahul@gmail.com", None, "FEST20", None, 3200.75),
    (1004, "Neha Gupta", None, "9123456780", None, None, 875.00),
    (1005, "Arjun Mehta", "arjun@gmail.com", None, None, "Bangalore", 5400.00)
]

df = spark.createDataFrame(data, schema)
df.display()
```

## What is coalesce()?

The **coalesce()** function returns the first non-null value from the given list of columns.

Syntax:

**coalesce(column1, column2, column3, ...)**

returns first available value

Example -1 replace missing coupon code

```python
from pyspark.sql.functions import coalesce , col , lit
```

```python
df.withColumn("final_coupon",coalesce(col("coupon_code"),lit("No coupon applied"))).display()
```

```python
df.withColumn("Primary_Contact",
              coalesce(col("phone_number"),col("email"),lit("Contact not Available"))
              ).display()
```

```python
df_cleaned = df \
    .withColumn("email", coalesce(col("email"), lit("No Email Provided"))) \
    .withColumn("phone_number", coalesce(col("phone_number"), lit("No Phone Provided"))) \
    .withColumn("shipping_address", coalesce(col("shipping_address"), lit("Address Not Provided"))) \
    .withColumn("coupon_code", coalesce(col("coupon_code"), lit("No Coupon Applied")))

df_cleaned.display()
```

⚠️ VERY IMPORTANT - Interview Question

There are TWO different coalesce functions in PySpark.

1️⃣ Column-level functions.coalesce()
→ Used for handling NULL values.

2️⃣ DataFrame-level df.coalesce()
→ Used to reduce partitions.
